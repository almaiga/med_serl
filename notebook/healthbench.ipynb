# HealthBench Dataset
Loading the OpenAI HealthBench dataset from Hugging Face

```python
from datasets import load_dataset

# Load the HealthBench dataset
dataset = load_dataset("openai/healthbench")

# Display dataset info
print(dataset)
```

```python
# Explore the dataset structure
for split in dataset:
    print(f"\n{split}: {len(dataset[split])} examples")
    print(f"Columns: {dataset[split].column_names}")
```
